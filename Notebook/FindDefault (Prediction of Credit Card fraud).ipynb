Next, we will divide the credit card dataset into training and testing sets using a 70-30 split with the train_test_split() function.
The train_test_split() function from scikit-learn is a convenient tool for partitioning a dataset into training and test subsets.

**Parameters**:
- X: Feature matrix.
- Y: Target variable.    
**test_size**: Specifies the proportion of the dataset used for testing. Here, it is set to 0.3, allocating 30% of the data to the test set.  
**random_state**: A fixed seed for the random number generator is provided to ensure the results are reproducible.

```python
from sklearn.model_selection import train_test_split

(train_X, test_X, train_Y, test_Y) = train_test_split(X,Y, test_size=0.3, random_state=42)

print("Shape of the training dataset train_X: ", train_X.shape)
print("Shape of the testing dataset test_X: ", test_X.shape)
```

# Applying Machine Learning Algorithms to the Credit Card Dataset  

We will evaluate various machine learning algorithms to identify the best model for our binary classification task.  
The objective is to predict one of two class labels. Our focus will be on assessing algorithms like **Random Forest** and **Decision Tree** to determine the most suitable solution.  

Our strategy involves building and comparing **Random Forest** and **Decision Tree** classifiers to identify the most effective model for this task.  

---

## Decision Tree Algorithm  

A **Decision Tree** is a supervised learning algorithm that splits the dataset into subsets based on feature values, creating a tree-like structure of decisions.  

- **Internal Nodes**: Represent features.  
- **Branches**: Represent decision rules.  
- **Leaf Nodes**: Represent class labels (classification) or continuous values (regression).  

### Key Concepts:  
1. **Recursive Splitting**: The tree is built by recursively splitting the data at each node based on a feature that results in the most significant information gain (classification) or variance reduction (regression).  
2. **Splitting Criterion**: Metrics like Gini impurity, entropy, or variance are used to determine the best feature for splitting at each node.  

---

## Random Forest Algorithm  

**Random Forest** is an ensemble learning method that combines multiple decision trees to improve accuracy and prevent overfitting.  

### How It Works:  
1. **Bootstrapping**: Each tree is trained on a random subset of the original dataset (with replacement), known as bootstrapping.  
2. **Random Feature Selection**: At each split in a decision tree, a random subset of features is considered for the best split, reducing correlation between trees.  
3. **Final Prediction**:  
   - For classification, the output is based on the majority vote from all trees.  
   - For regression, the output is the average prediction from all trees.  

This ensemble approach enhances model robustness and ensures better generalization to unseen data.  


# Training and Evaluation of models on the dataset


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier()

# Random Forest
random_forest = RandomForestClassifier(n_estimators=100)
```

In this case, we initialize a **RandomForestClassifier** with 100 trees in the ensemble.  
While increasing the number of trees generally enhances the model's performance, it can also lead to longer training times.  


```python
# Decision Tree Classifier
decision_tree.fit(train_X, train_Y)
predictions_data = decision_tree.predict(test_X)

decision_tree_score = decision_tree.score(test_X, test_Y) * 100

# Printing the score of the classifier
print("Decision Tree Score is: ", decision_tree_score)
```

```python
# Random Forest
random_forest.fit(train_X, train_Y)
predictions_data_rf = random_forest.predict(test_X)
```

```python
random_forest_score = random_forest.score(test_X, test_Y) * 100

# Printing the score of the Random Forest Classifiers
print("Random Forest Score is: ", random_forest_score)
```

```python
# Printing the scores of the both classifiers
print("Decision Tree: ", round((decision_tree_score),2))
print("Random Forest: ", round((random_forest_score),2))
```


**Random Forest** outperformed the **Decision Tree** in our evaluation.

```python
print("Evaluation of Decision Tree Model: ")

print("Accuracy:",round(accuracy_score(test_Y,predictions_data),4))
print("Precision:",round(precision_score(test_Y,predictions_data),4))
print("recall_score:",round(recall_score(test_Y,predictions_data),4))
print("F1-Score:",round(f1_score(test_Y,predictions_data),4))
```

### Class Imbalance  

- The **Random Forest model** often outperforms Decision Trees, particularly when addressing class imbalances.  
- In this case, genuine transactions make up over **99% of the dataset**, while fraudulent transactions account for only **0.17%**.  
- Training without handling this imbalance can result in biased predictions.  
- While the model might seem accurate overall, it may fail to effectively detect the **minority class** (fraudulent transactions) and could struggle in real-world applic


```python
# We will use **SMOTE** (Synthetic Minority Oversampling Technique).  
# SMOTE is a data augmentation method designed for the minority class.

from imblearn.over_sampling import SMOTE

X_resampled, Y_resampled = SMOTE().fit_resample(X,Y)

print("Resampled shape of X: ",X_resampled.shape)
print("Resampled shape of Y: ",Y_resampled.shape)
```

```python
from collections import Counter
value_counts = Counter(Y_resampled)
print(value_counts)
```

```python
(train_X, test_X, train_Y, test_Y) = train_test_split(X_resampled, Y_resampled, test_size=0.3, random_state=42)
```

```python
rf_resampled = RandomForestClassifier(n_estimators=100)
rf_resampled.fit(train_X, train_Y)
```

#### - Since **Random Forest** outperformed the **Decision Tree**, we will apply it to the resampled data.  


```python
predictions_resampled = rf_resampled.predict(test_X)
random_forest_score_resampled = rf_resampled.score(test_X, test_Y) * 100
print(random_forest_score_resampled)
```

# Evaluation Matrics of Resampled data

Our model shows a substantial improvement in performance over the previous **Random Forest** classifier, which did not utilize oversampling. By implementing techniques to address class imbalance, we achieved an accuracy rate of over **99%**.
